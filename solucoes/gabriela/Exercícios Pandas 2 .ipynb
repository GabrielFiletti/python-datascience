```python
import pandas as pd
```

## 1 - Leia os dados train.csv usando pandas.

```python
df = pd.read_csv("https://raw.githubusercontent.com/advinstai/python-datascience/master/Exercicios/titanic/train.csv")
```

## 2 - Remova a coluna com nomes de pessoas.

```python
print(df.columns)
df = df.drop(['Name'],axis=1)
print(df.columns)
```

## 3 - A coluna Cabin contém uma letra e um número. Crie uma nova coluna chamada deck usando apenas a letra de cada registro presente na coluna Cabin.

```python
df["Deck"] = [x[0] if not pd.isnull(x) else x for x in df["Cabin"]]
df[["Cabin","Deck"]]
```

## 4 - Substitua os valores ausentes pela média de valores da coluna

```python

df.fillna(df.mean())   
```

## 6 - Crie um CSV com uma coluna nome_cidade_de_embarque: C = Cherbourg, Q = Queenstown, S = Southampton e mais uma coluna chamada "população" com valores hipotéticos, faça um merge entre o CSV train.csv e esse novo CSV com os nomes das cidades. Implemente um filtro de passageiros por população da cidade de embarque

```python
df2 = pd.DataFrame({'sigla':['C', 'Q', 'S'], 'nome_cidade_de_embarque': ['Cherbourg', 'Queenstown', 'Southampton'],
                    'populacao': [85000, 15850, 253600] })
df2.to_csv("embarque.csv",index=False)
```

```python
mergedDf = pd.merge(df,df2,how='left', left_on="Embarked",right_on="sigla")
```

```python
df = df.merge(df2, how='left', left_on="Embarked",right_on="sigla")
```

```python
# Se tabela não possui coluna Name, retorna Id dos passageiros
def passenger_filter(cidade_embarque):
    try: 
        print(df.loc[df['nome_cidade_de_embarque']==cidade_embarque,"Name"].to_string(index=False))
    except:
        print(df.loc[df['nome_cidade_de_embarque']==cidade_embarque,"PassengerId"].to_string(index=False))

cidade = input("Insira cidade de embarque: ")
passenger_filter(cidade)
```
