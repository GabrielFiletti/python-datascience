##  Contar a quantidade de elementos nulos por coluna.

```python
# Identifica as linhas com horsepower = NaN
# Identifica as linhas que horsepower=NaN com a fun√ß√£o isnull()
# A fun√ß√£o shape retorna uma tupla com o n√∫mero de linhas e colunas (linhas x colunas) do df
# Pegando o primeiro item (shape[0]) obtemos o n√∫mero de linhas desta parte do df

df[df['horsepower'].isnull()].shape[0]
```

```python
# Forma mais simples:
df['horsepower'].isnull().sum()
```

## Criar um dataframe eliminando as linhas com valores nulos.

```python
newdf = df.dropna()
newdf
```

## Criar um dataframe transformando os valores nulos para zero.

```python
df_zero = df.fillna(0)
```

```python
df_zero[df_zero['horsepower'] == 0]
```

## Criar um dataframe transformando os valores nulos para a m√©dia da coluna.

```python
df['horsepower'].mean()
```

```python
# Conferindo:
df.describe()
```

```python
# Substituindo as c√©lulas com NaN pela m√©dia
# Cria um novo dataframe substituindo os NaN pela m√©dia
df_mean = df.fillna(df['horsepower'].mean())
```

```python
# Apresenta na tela a parte do dataframe com horsepower igual a m√©dia
df_mean[df_mean['horsepower'] == df_mean['horsepower'].mean()]
```
