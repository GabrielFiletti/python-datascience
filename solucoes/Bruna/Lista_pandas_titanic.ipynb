* Leia os dados train.csv usando pandas

```python
import pandas as pd

#abrindo o arquivo por html
url='https://raw.githubusercontent.com/advinstai/python-datascience/master/Exercicios/titanic/train.csv'
from_url = pd.read_table(url, sep=',')  #le a tabela e separa por virgula

from_url
```

* Remova a coluna com nomes de pessoas

```python
from_url.drop(columns=['Name'])
```

* A coluna Cabin contÃ©m uma letra e um nÃºmero. Crie uma nova coluna chamada deck usando apenas a letra de cada registro presente na coluna Cabin.

```python
#Criando um novo dataframe a partir de from_url
new_data = from_url

#criando uma nova coluna espelhada a partir da coluna Cabin
new_data["deck"] = from_url["Cabin"]

#aplicando ao dataframe, coluna deck, dados string tal que tudo o que nao for letra(replace("\d+","")) recebe vazio
new_data.deck.str.replace("\d+","").dropna()
```

* Substitua os valores ausentes pela mÃ©dia de valores da coluna 

```python
new_data = new_data.fillna(new_data.mean())
new_data.head()
```

* Os valores da coluna "embarked" sao os seguintes: C = Cherbourg, Q = Queenstown, S = Southampton.  Crie um CSV com essas uma coluna nome_cidade_de_embarque: C = Cherbourg, Q = Queenstown, S = Southampton e mais uma coluna chamada populaÃ§Ã£o com valores hipotÃ©ticos, faÃ§a um merge entre o CSV train.csv e esse novo CSV com os nomes das cidades. Implemente um filtro de passageiros por populaÃ§Ã£o da cidade de embarque

```python
#dicionario
dados_csv = {'city': 'C', 'population': 10000 }, {'city': 'Q', 'population': 50000 }, {'city': 'S', 'population': 70000 }

#transformando o filtro em DataFrame
dfcsv = pd.DataFrame(data = dados_csv)

#Realizando o merge entre os dois DataFrame
df_final = pd.merge(from_url, dfcsv, left_on = 'Embarked' , right_on = 'city', how = 'inner')

#filtro

l = df_final.loc[(df_final["population"] > 50000)]

l
```
