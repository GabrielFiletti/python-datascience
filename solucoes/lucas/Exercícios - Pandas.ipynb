```python
#2. Encontre o nome da empresa de carros mais cara.
url = 'https://raw.githubusercontent.com/advinstai/python-datascience/master/Exercicios/carros.csv'
data_from_url = pd.read_csv(url)
dfObj = pd.DataFrame(data_from_url)

cars_Company = dfObj.groupby('company')
priceDf = cars_Company['company','price'].max()
priceDf
```

```python
#3. Mostre a média de preços.
url = 'https://raw.githubusercontent.com/advinstai/python-datascience/master/Exercicios/carros.csv'
data_from_url = pd.read_csv(url)
dfObj = pd.DataFrame(data_from_url)

cars_CompanyPrice = dfObj.groupby('company')
pricesDf = cars_CompanyPrice['company','price'].mean()
pricesDf
```

```python
#4. Mostre o valor mais alto e mais baixo de horsepower.
url = 'https://raw.githubusercontent.com/advinstai/python-datascience/master/Exercicios/carros.csv'
data_from_url = pd.read_csv(url)
dfObj = pd.DataFrame(data_from_url)


print('Maior preço (HorsePower):')
highPriceHP = dfObj[['horsepower', 'price']][dfObj.price==dfObj['price'].max()]
print(highPriceHP)

print('Menor preço (HorsePower):')
lowPriceHP = dfObj[['horsepower', 'price']][dfObj.price==dfObj['price'].min()]
print(lowPriceHP)
```

```python
#5. Mostre a informação de todos os carros da marcar Toyota.
data_from_url = pd.read_csv(url)
dfObj = pd.DataFrame(data_from_url)

cars_Toyota = dfObj.groupby('company')
toyotaDf = cars_Toyota.get_group('toyota')
cars_Toyota
toyotaDf
```

```python
#6. Faça uma função que filtra os carros por cilindro e salva em outro arquivo CSV.
url = 'https://raw.githubusercontent.com/advinstai/python-datascience/master/Exercicios/carros.csv'
data_from_url = pd.read_csv(url)

dfObj = pd.DataFrame(data_from_url)

def carsByCylinders(number):
    cars_Cylinders = dfObj.groupby('num-of-cylinders')
    cylinders_number = cars_Cylinders.get_group(number)
    cylinders_number.to_csv(r'problema6')
    return cylinders_number

number = 'six'
carsByCylinders(number)
```
