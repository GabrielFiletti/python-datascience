# Lista de Exercícios 1
### 1. Do arquivo carros, imprima as primeiras e as últimas cinco linhas. Faça uma versão que abre o arquivo a partir do csv e outra que abre a partir da url.

```python
def printheadandtail(path,size=5):
    data = pd.read_csv(path)
    out = pd.concat([data.head(size),data.tail(size)])
    return out
```

```python
file = 'https://raw.githubusercontent.com/advinstai/python-datascience/master/Exercicios/carros.csv'
printheadandtail(file)
```

```python
printheadandtail(file)
```

### 2. Encontre o nome da empresa de carros mais cara

```python
def expensive(path):
    data = pd.read_csv(path)
    mp = np.max(data['price'])
    return data[data['price']== mp]
```

```python
expensive(file)
```

### 3. Mostre a média de preços

```python
def meanprice(path):
    data = pd.read_csv(path)
    return data['price'].mean()
```

```python
meanprice(file)
```

### 4. Mostre o valor mais alto e mais baixo de horsepower

```python
def extremehp(path):
    data = pd.read_csv(path)
    return data['horsepower'].max(),data['horsepower'].min()
```

```python
extremehp(file)
```

### 5. Mostre a informação de todos os carros da marcar toyota

```python
def showcatalog(path,company):
    data = pd.read_csv(path)
    return data[data['company'] == company]
```

```python
showcatalog(file,'toyota')
```

### 6. Faça uma função que filtra os carros por cilindro e salva em outro arquivo CSV

```python
def filtercsv(path,par,num):
    data = pd.read_csv(path)
    df = data[data[par] == num]
    df.to_csv(r'./output.csv')
```

```python
filtercsv(file,'num-of-cylinders','four')
```
