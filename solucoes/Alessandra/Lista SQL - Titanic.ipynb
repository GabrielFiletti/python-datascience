1. Leia os dados train.csv usando pandas

```python
import sqlite3
import pandas as pd

dfTitanic = pd.read_csv('https://raw.githubusercontent.com/advinstai/python-datascience/master/Exercicios/titanic/train.csv',",")

display(dfTitanic)
```

2. Remova a coluna com nomes de pessoas

```python
dfTitanic = dfTitanic.drop('Name',axis=1)
display(dfTitanic)
```

3. A coluna Cabin contÃ©m uma letra e um nÃºmero. Crie uma nova coluna chamada deck usando apenas a letra de cada registro presente na coluna Cabin.

```python
dfTitanic['deck'] = dfTitanic['Cabin'].astype(str).str[0]
display(dfTitanic)
```

4. Substitua os valores ausentes pela mÃ©dia de valores da coluna

```python
dfTitanic = dfTitanic.fillna(dfTitanic.mean())
display(dfTitanic)
```

5 Os valores da coluna "embarked" sao os seguintes: C = Cherbourg, Q = Queenstown, S = Southampton

6 Crie um CSV com essas uma coluna nome_cidade_de_embarque: C = Cherbourg, Q = Queenstown, S = Southampton e mais uma coluna chamada populaÃ§Ã£o com valores hipotÃ©ticos, faÃ§a um merge entre o CSV train.csv e esse novo CSV com os nomes das cidades. Implemente um filtro de passageiros por populaÃ§Ã£o da cidade de embarque

```python
import numpy as np
x = {'C':"Cherbourg", 'Q':"Queenstown", 'S':"Southampton"}
y = {'C':150030, 'Q':270130, 'S':305200}
dfTitanic['nome_cidade_de_embarque'] = dfTitanic['Embarked']
dfTitanic['populacao'] = dfTitanic['Embarked']
dfTitanic = dfTitanic.replace({"nome_cidade_de_embarque": x})
dfTitanic = dfTitanic.replace({"populacao": y})
dfTitanic.to_csv('Titanic.csv')

dfTrain = pd.read_csv('https://raw.githubusercontent.com/advinstai/python-datascience/master/Exercicios/titanic/train.csv')
dfMerge = pd.merge(dfTrain, dfTitanic)

dfFilter = dfMerge[(dfMerge['populacao']==150030)]
display(dfFilter)
```
