## 1. Leia os dados train.csv usando pandas

```python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/advinstai/python-datascience/master/Exercicios/titanic/train.csv"
url1 = pd.read_table(url, sep=',')

display(url1.head(5))
display(url1.tail(5))
print(url1)
```

## 2. Remova a coluna com nomes de pessoas

```python
url2 = url1.drop(columns="Name")
print(url2)
```

## 3. A coluna Cabin contém uma letra e um número. Crie uma nova coluna chamada deck usando apenas a letra de cada registro presente na coluna Cabin.

```python
url30 = url2

url30["Cabin"] = url30["Cabin"].fillna(0)
print(url30)

url30["deck"] = url30["Cabin"]
display(url30)

url30.deck = url30.deck.str.replace('\d+', '')
#url3["deck"] = url3["deck"].fillna(0)

print(url30)
```

## 4. Substitua os valores ausentes pela média de valores da coluna

```python
url41 = url30
display(url41)

url42 = url41.fillna(url41.mean())

url42["deck"] = url42["deck"].replace(np.nan, 0)
display(url42)
```

## 5. Os valores da coluna "embarked" sao os seguintes: C = Cherbourg, Q = Queenstown, S = Southampton.
## Crie um CSV com essas uma coluna nome_cidade_de_embarque: C = Cherbourg, Q = Queenstown, S = Southampton e mais uma coluna chamada população com valores hipotéticos, faça um merge entre o CSV train.csv e esse novo CSV com os nomes das cidades. Implemente um filtro de passageiros por população da cidade de embarque

```python
url50 = url30
url50["city"] = url30["Embarked"]
display(url50)
```

```python
dictionary = {"C":"Cherbourg", "Q":"Queenstown", "S":"Southampton"}
url51 = url50.replace({"city": dictionary})
display(url51)

url51.to_csv("titanic.csv")
```

```python
dic2 = [{'city': 'Cherbourg', 'population': 36550},
         {'city': 'Queenstown',  'population': 22365},
         {'city': 'Southampton',  'population': 9999}]

dfDic = pd.DataFrame(dic2)
display(dfDic)

dfDic.to_csv("dfDic.csv")

```

```python
dfnova = pd.merge(url51, dfDic, left_on = "city", right_on = "city", how = "outer")
display(dfnova)
```

```python
display(dfnova.loc[dfnova['population'] >= 10000])
```
